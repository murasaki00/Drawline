```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

#Load an color image 
filename = 'lena.jpg' #input name of jpeg file
img=cv2.imread(filename)   

image_size=480 #input image size
image_devide=45 #input interval of lines

adj_0 =0
adj_1 =0

if img.shape[1]>img.shape[0]:
	resize_img=cv2.resize(img,image_size*img.shape[1]//img.shape[0],image_size)
	adj_1 = 1
	
else:
  resize_img=cv2.resize(img,(image_size, image_size*img.shape[0]//img.shape[1]))
  adj_0 = 1

for i in range(0, resize_img.shape[1]//image_devide + adj_1):

	for j in range(0,resize_img.shape[0]):

		resize_img[j][image_devide*i]=0

for i in range(0,resize_img.shape[1]):

	for j in range(0, resize_img.shape[0]//image_devide + adj_0):

		resize_img[image_devide*j][i]=0

#print('height',round(resize_img.shape[0]/image_devide))
#print('width',round(resize_img.shape[1]/image_devide))


cv2_imshow(resize_img)
```
